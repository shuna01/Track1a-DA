```python
print('Welcome to the Rock, Paper, Scissors Game!')

start_game = True

while start_game:
    play = True
    while play:
        ask_player1_play = input('Player 1, please enter play to start or quit to exit the game: ')
        if ask_player1_play == 'play':
            play = False
            print('Let the fun begin!')
        elif ask_player1_play == 'quit':
            print('End Game. See you next time!')
            play = False
            start_game = False
            break
        else:
            print('You have entered an invalid response. Please enter play to start or quit to exit the game: ')
    if ask_player1_play == 'quit':
        start_game = False
        break
        
    game = True
    while game:
        invalidp1input = True 
        while invalidp1input:
            player1_input = input('Player 1, please enter rock, paper or scissors: ')
            if player1_input != 'rock' and player1_input != 'paper' and player1_input != 'scissors':
                print('Invalid entry. Please enter rock, paper or scissors only: ')
            else:
                invalidp1input = False

        invalidp2input = True
        while invalidp2input:
            player2_input = input('Player 2, please enter rock, paper or scissors: ')
            if player2_input != 'rock' and player2_input != 'paper' and player2_input != 'scissors':
                print('Invalid entry. Please enter rock, paper or scissors only: ')
            else:
                invalidp2input = False

        if player1_input == 'rock' and player2_input == 'paper':
            print('Player 2 wins!')
        elif player1_input == 'rock' and player2_input == 'scissors':
            print('Player 1 wins!')
        elif player1_input == 'paper' and player2_input == 'rock':
            print('Player 1 wins!')
        elif player1_input == 'paper' and player2_input == 'scissors':
            print('Player 2 wins!')
        elif player1_input == 'scissors' and player2_input == 'rock':
            print('Player 2 wins!')
        elif player1_input == 'scissors' and player2_input == 'paper':
            print('Player 1 wins!')
        else:
            print('It is a draw!')
    
        new_game = True
        while new_game:
            ask_player1_newgame = input('Would you like to play again? Please enter yes or no: ')
            if ask_player1_newgame == 'yes':
                new_game = False
            elif ask_player1_newgame == 'no':
                print('Thanks for playing. See you next time!')
                new_game = False
            else:
                print('Invalid response. Please enter yes or no: ')
        if ask_player1_newgame == 'yes':
            game = True
        if ask_player1_newgame == 'no':
            game = False
            start_game = False
            break
```
